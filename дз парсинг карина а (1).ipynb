### 2) Сделать список не по первой странице, а по первым нескольким страницам хабра

Здесь я создаю словарь, в котором будут зраниться датафреймы страниц            

Чтобы обратиться к определенной стронице (со 2 до 5 включительно), запросите:

parsed_dataframes['parsed_i']             
где i - номер страницы

```python
parsed_dataframes = {}
count = 0
for q in range(2, 6):
    count += 1
    req = r.get('https://habr.com/ru/feed/page' + f'{q}/', headers = headers)
    html_doc = req.text
    soup = BeautifulSoup(html_doc, 'html.parser') 
    
    titles = []
    names = []
    time = []
    theme = []
    links = []
    likes = []
    comms = []
    
    parsed_dataframes[f'parsed_{q}'] = pd.DataFrame()
    
    for i in soup.find_all('article'):
        theme1 = []
        
        try:
            titles.append(i.find('h2').text)
        except AttributeError:
            titles.append(' ')
        
        try:
            names.append(i.find('span', class_="tm-user-info tm-article-snippet__author").text.split(' ')[0])
        except AttributeError:
            names.append(' ')
        
        try:
            for j in i.find_all(class_ = 'tm-publication-hub__link'):
                theme1.append(j.text)
            theme.append(theme1)
        except AttributeError:
            theme.append(' ')
            
        try:
            time.append(i.find('span', class_ = 'tm-article-reading-time__label').text)
        except AttributeError:
            time.append(' ')
                     
        try:
            links.append('https://habr.com' + i.find('a', class_ = 'tm-title__link').get('href'))
        except AttributeError:
            links.append(' ')
        
        comms.append(i.find('span', class_ = 'tm-article-comments-counter-link__value').text)   
        
        like_element = (
            i.find('span', class_='tm-votes-meter__value tm-votes-meter__value_appearance-article tm-votes-meter__value_rating tm-votes-meter__value') or
            i.find('span', class_='tm-votes-meter__value tm-votes-meter__value_negative tm-votes-meter__value_appearance-article tm-votes-meter__value_rating tm-votes-meter__value') or
            i.find('span', class_='tm-votes-meter__value tm-votes-meter__value_positive tm-votes-meter__value_appearance-article tm-votes-meter__value_rating tm-votes-meter__value')
        )
        likes.append(like_element.text if like_element else ' ')                

    parsed_dataframes[f'parsed_{q}']['titles'] = titles
    parsed_dataframes[f'parsed_{q}']['names'] = names
    parsed_dataframes[f'parsed_{q}']['time'] = time
    parsed_dataframes[f'parsed_{q}']['theme'] = theme
    parsed_dataframes[f'parsed_{q}']['links']= links
    parsed_dataframes[f'parsed_{q}']['comms'] = comms
    parsed_dataframes[f'parsed_{q}']['likes'] = likes
```

```python
for q in range(2, 6):
    for i in range(0, parsed_dataframes[f'parsed_{q}'].shape[0]):
        parsed_dataframes[f'parsed_{q}']['theme'][i] = (', ').join(parsed_dataframes[f'parsed_{q}']['theme'][i])
```

### 3) Проследовать по ссылкам в профили авторов и узнать их имя, должность, карму, рейтинг, число публикаций и комментариев. Добавить эту информацию к инфе про статьи

в каждый датафрейм из словаря добавляются новые колонки:         

*real_names* - имена авторо_к  
*empl* - должность  
*carma* - карма  
*rating* - рейтинг  
*num* - количество публикаций  
*comments* - количество комментариев

```python
for j in parsed_dataframes:
    real_names = []
    empl = []
    carma = []
    rating = []
    num = []
    comments = []

    for i in parsed_dataframes[j]['names']:

        if i == ' ':

            real_names.append(' ')
            empl.append(' ')
            carma.append(' ')
            rating.append(' ')
            num.append(' ')
            comments.append(' ')
        else:
            req = r.get('https://habr.com/ru/users/' + f'{i}' + '/', headers = headers)
            html_doc = req.text
            soup = BeautifulSoup(html_doc, 'html.parser') 
            try:
                real_names.append(soup.find('span', class_ = 'tm-user-card__name tm-user-card__name_variant-profile tm-user-card__name').text)
            except AttributeError:
                real_names.append(' ')
            try:
                empl.append(soup.find('p', class_ = 'tm-user-card__short-info tm-user-card__short-info_variant-profile tm-user-card__short-info').text)
            except AttributeError:
                empl.append(' ')
            try:
                carma.append(soup.find(class_ = 'tm-karma__votes_positive tm-karma__votes').text)
            except AttributeError:
                carma.append(' ')
            try:
                rating.append(soup.find(class_ = 'tm-votes-lever__score-counter tm-votes-lever__score-counter_rating tm-votes-lever__score-counter').text)
            except AttributeError:
                rating.append(' ')
            try:
                num.append(soup.find(class_ = 'tm-tabs__tab-counter').text)
            except AttributeError:
                num.append(' ')
            for q in soup.find_all('span', class_="tm-tabs__tab-item"):
                if q.find('a').text.split(' ')[0] == 'Комментарии':
                    comments.append(q.find('a').text.split(' ')[1])
    parsed_dataframes[j]['real_names'] = real_names
    parsed_dataframes[j]['empl'] = empl
    parsed_dataframes[j]['carma'] = carma
    parsed_dataframes[j]['rating'] = rating
    parsed_dataframes[j]['num'] = num
    parsed_dataframes[j]['comments'] = comments
```
